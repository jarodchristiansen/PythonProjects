```
cities = ['Yangon', 'Naypyitaw', 'Mandalay']
index = ['Gross Income', 'Males', 'Females', 'Home/Life', 'Electronics', 'Sports/Travel', 'Food', 'Fashion', 'Health']


city_data = pd.DataFrame(index=index)



def lister(df, df2, df3, new_list):
  income = df['gross income'].sum()
  males = df2['gross income'].sum()
  females = df3['gross income'].sum()
  nums = df['Product line'].value_counts()
  home = df[df['Product line'].str.contains('home')]
  home = home['gross income'].sum()

  electronics = df[df['Product line'].str.contains('electronics')]
  electronics = electronics['gross income'].sum()

  sports = df[df['Product line'].str.contains('sports')]

  food = df[df['Product line'].str.contains('food')]

  fashion = df[df['Product line'].str.contains('fashion')]

  health = df[df['Product line'].str.contains('health')]
  
  new_list.append(income)
  new_list.append(males)
  new_list.append(females)
  new_list.append(nums[0] + str(home))
  new_list.append(nums[1] + str(electronics))
  new_list.append(nums[2] + str(sports))
  new_list.append(nums[3] + str(food))
  new_list.append(nums[4] + str(fashion))
  new_list.append(nums[5] + str(health))
  print(new_list)

yangon_list = []
naypyitaw_list = []
mandalay_list = []



lister(yangon, yangon_male, yangon_female, yangon_list)
lister(naypyitaw, naypyitaw_male, naypyitaw_female, naypyitaw_list)
lister(mandalay, mandalay_male, mandalay_female, mandalay_list)
```
